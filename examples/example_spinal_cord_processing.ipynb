## Stick and Ball

```python
ball = three_dimensional_models.E3Ball()
stick = three_dimensional_models.I1Stick()

ball_and_stick = three_dimensional_models.PartialVolumeCombinedMicrostrukturModel(
    [stick, ball], parameter_links=[], optimise_partial_volumes=True)
```

```python
x0 = ball_and_stick.parameters_to_parameter_vector(I1Stick_1_lambda_par=2,
                                                   I1Stick_1_mu=[0, 0],
                                                   E3Ball_1_lambda_iso=1,
                                                   partial_volume_0=.5
                                                   )

results_ball_and_stick = np.zeros((ROI_data.shape[0], len(ball_and_stick.parameter_cardinality) + 1))
for i, data_ in enumerate(ROI_data):
    res = optimize.minimize(
        ball_and_stick.objective_function, x0, (bvals_3d, bvecs_3d, data_),
        bounds=ball_and_stick.bounds_for_optimization, options={'disp': True})
    results_ball_and_stick[i] = res.x
```

```python
plt.scatter(results_ball_and_stick[:, -1], ROI_fractions)
plot([0, 1], [0, 1])
ylim(0, .6)
xlim(0, .6)

pr = pearsonr(results_ball_and_stick[:, -1], ROI_fractions)
text(.35, .06, 'PearsonR: '+str(np.round(pr[0], 3)), fontsize=15)
```

## Multi-Compartment Spherical Mean Technique

```python
stick_mean = three_dimensional_models.I1StickSphericalMean()
zeppelin_mean = three_dimensional_models.E4ZeppelinSphericalMean()
```

```python
parameter_links = [
    (  # tortuosity assumption
        zeppelin_mean, 'lambda_perp',
        utils.T1_tortuosity, [
            (None, 'partial_volume_0'),
            (stick_mean, 'lambda_par')
        ]
    ),
    (  # equal parallel diffusivities
         zeppelin_mean, 'lambda_par',
         utils.parameter_equality, [
            (stick_mean, 'lambda_par')
        ]
     )
]
```

```python
mc_smt = three_dimensional_models.PartialVolumeCombinedMicrostrukturModel(
    [stick_mean, zeppelin_mean], parameter_links=parameter_links, optimise_partial_volumes=True)
```

```python
bvalue_ranges = ((0, 10), (30, 50), (180, 200), (1500, 1700), (5000, 7000))
shell_indices, shell_bvals = utils.define_shell_indices(bvals_3d, bvalue_ranges)

number_of_shells = len(bvalue_ranges)
data_spherical_means = np.zeros((ROI_data.shape[0], number_of_shells - 1))
for i, data_ in enumerate(ROI_data):
    data_spherical_means[i] = spherical_mean.estimate_spherical_mean_multi_shell(
                                data_, bvecs_3d, shell_indices, sh_order=10)
```

```python
x0 = mc_smt.parameters_to_parameter_vector(I1StickSphericalMean_1_lambda_par=1.3,
                                           partial_volume_0=0.30)

results_mc_smt = np.zeros((ROI_data.shape[0], 2))
for i, data_spherical_mean_ in enumerate(data_spherical_means):
    res = optimize.minimize(mc_smt.objective_function, x0,
                            (shell_bvals[1:], bvecs_3d[:4], data_spherical_mean_),
                            bounds=mc_smt.bounds_for_optimization, options={'disp': True})
    results_mc_smt[i] = res.x
```

```python
plt.scatter(results_mc_smt[:, -1], ROI_fractions)
plot([0, 1], [0, 1])
ylim(0, 1)
xlim(0, 1)

pr = pearsonr(results_mc_smt[:, -1], ROI_fractions)
text(.6, .06, 'PearsonR: '+str(np.round(pr[0], 3)), fontsize=15)
```

```python
mcsmt_image = np.zeros_like(mask1plus) + 0.
mcsmt_image[mask1plus] = results_mc_smt[:, -1]
imshow(mcsmt_image[8:55, 10:44].T, origin=True, cmap='Greys_r')
colorbar()
```

## Apparent Fiber Density

```python
from dipy.reconst.csdeconv import auto_response
from dipy.core.gradients import gradient_table
from dipy.reconst.csdeconv import ConstrainedSphericalDeconvModel
from dipy.data import get_sphere
sphere = get_sphere('symmetric724')

highest_bval_and_b0_mask = np.any([bvals_3d < 10, bvals_3d > 6000], axis=0)
masked_bvals = bvals_3d[highest_bval_and_b0_mask]
masked_bvecs = bvecs_3d[highest_bval_and_b0_mask]
masked_gtab = gradient_table(bvals=masked_bvals, bvecs=masked_bvecs)

masked_data = ROI_data[:, highest_bval_and_b0_mask]
```

```python
response, ratio = auto_response(masked_gtab, masked_data[:, None, None], roi_radius=1000, fa_thr=0.4)
csd_model = ConstrainedSphericalDeconvModel(masked_gtab, response)
csd_fit = csd_model.fit(masked_data)
arg_par = argmin(np.linalg.norm(sphere.vertices - np.r_[0, 0, 1], axis=1))
csd_odf = csd_fit.odf(sphere)
csd_amplitudes = csd_odf[:, arg_par]
```

```python
scatter(csd_amplitudes, ROI_fractions)
pr = pearsonr(csd_amplitudes, ROI_fractions)
text(2.25, .32, 'PearsonR: '+str(np.round(pr[0], 3)), fontsize=15)
```

## White Matter Tract Integrity

```python
from dipy.core.gradients import gradient_table
from dipy.reconst import dki

gtab = gradient_table(bvals=bvals_3d, bvecs=bvecs_3d)

dki_model = dki.DiffusionKurtosisModel(gtab)
dki_fit = dki_model.fit(ROI_data)

# radial (perpendicular) kurtosis is highest
k_max = dki_fit.rk(max_kurtosis=10)
dki_fintra = k_max / (k_max + 3)
```

```python
# without dispersion
scatter(dki_fintra, ROI_fractions)
plot([0, 1], [0, 1])
ylim(0, 1)
xlim(0, 1)
pr = pearsonr(dki_fintra, ROI_fractions)
text(.58, .05, 'PearsonR: '+str(np.round(pr[0], 4)), fontsize=15)
```
