# Bar-Schir et al experiments

Bar-Shir A, Avram L, Özarslan E, Basser PJ, Cohen Y (2008) The effect of the diffusion time and pulse gradient duration ratio on the diffraction pattern and the structural information estimated from q-space diffusion MR: Experiments and simulations. J Magn Reson 194:230–236. doi: 10.1016/j.jmr.2008.07.009

```python
figure(figsize=(15, 5))
disk = multiple_correlation_function.Disk()
delta = 2e-3
Q = linspace(0, 1500, 100) / 1e-2
G = signal_models.g_from_q(Q, delta)
gradient_echo_simulator = multiple_correlation_function.GradientStrengthEcho(
    gradient_strength=G,
    delta=delta,
    geometry=disk,
    length=20e-6 / 2
)

for Delta in r_[15, 30, 50, 100] * 1e-3:
    subplot(131)
    semilogy(
        Q * 1e-2,
        gradient_echo_simulator.attenuation_two_square_pulses(
            Time=Delta + 2 * delta
        )
    )
    ylim(10e-10, 1)
    subplot(132)
    semilogy(
        Q * 1e-2,
        gradient_echo_simulator.attenuation_stejkal_tanner(
            Time=Delta + 2 * delta
        )
    )
    ylim(10e-10, 1)
    subplot(133)
    semilogy(
        Q * 1e-2,
        gaussian_phase.cylinder_attenuation(
            gradient_echo_simulator.gradient_strength,
            Delta,
            gradient_echo_simulator.delta,
            gradient_echo_simulator.length * 2,
        )
    )
    ylim(10e-10, 1)

subplot(131)
title('MCF')
subplot(132)
title('NPA')
subplot(133)
title('GPA')
suptitle('Fig 2')
```

```python
figure(figsize=(15, 5))
disk = multiple_correlation_function.Disk()
delta = 32e-3
Q = linspace(0, 2500, 100) / 1e-2
G = signal_models.g_from_q(Q, delta) 
gradient_echo_simulator = multiple_correlation_function.GradientStrengthEcho(
    gradient_strength=G,
    delta=delta,
    geometry=disk,
)

for Delta in r_[42, 100, 300] * 1e-3:
    subplot(131)
    semilogy(
        Q * 1e-2,
        (
            .5 * (
            gradient_echo_simulator.attenuation_two_square_pulses(
                Time=Delta + 2 * delta, length=15e-6 / 2) +
            gradient_echo_simulator.attenuation_two_square_pulses(
                Time=Delta + 2 * delta, length=20e-6 / 2))
        )
    )
    ylim(10e-5, 1)
    subplot(132)
    semilogy(
        Q * 1e-2,
        (
            .5 * (
            gradient_echo_simulator.attenuation_stejkal_tanner(
                Time=Delta + 2 * delta, length=15e-6 / 2) +
            gradient_echo_simulator.attenuation_stejkal_tanner(
                Time=Delta + 2 * delta, length=20e-6 / 2))
        )
    )
    ylim(10e-5, 1)
    subplot(133)
    semilogy(
        Q * 1e-2,
        .5 * ( gaussian_phase.cylinder_attenuation(
            gradient_echo_simulator.gradient_strength, Delta, gradient_echo_simulator.delta, 15e-6 
        ) + gaussian_phase.cylinder_attenuation(
            gradient_echo_simulator.gradient_strength, Delta, gradient_echo_simulator.delta, 20e-6
        )
    )
    )
    ylim(10e-5, 1)

subplot(131)
title('MCF')
subplot(132)
title('NPA')
subplot(133)
title('GPA')
suptitle('Fig 7a')
```
